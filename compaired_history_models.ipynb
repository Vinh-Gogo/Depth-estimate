```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_cae = pd.read_csv('ALL_CODE\CAE_Resnet_All.csv').drop(columns=['Unnamed: 0'])
df_unet = pd.read_csv(f'Unet_Resnet_All.csv').drop(columns=['Unnamed: 0'])
# df_unet_plus = pd.read_csv('Unet3_Res_Den_His.csv').drop(columns=['Unnamed: 0'])
df_cae.shape, df_unet.shape
```

```python
def plot_comparison_training_history(len_cae, history_cae, len_unet, history_unet, len_unet_plus, history_unet_plus):

    plt.figure(figsize=(12, 6))
    print(f"Length of CAE history: {len_cae}")
    print(f"Length of UNet history: {len_unet}")
    print(f"Length of UNet Plus history: {len_unet_plus}")
    # Plotting Loss
    plt.subplot(1, 2, 1)
    plt.plot(range(len_cae), history_cae['train_loss'].values, label='CAE', linestyle='-', marker='x', markersize=3)
    plt.plot(range(len_unet), history_unet['train_loss'].values, label='UNet', linestyle='-', marker='x', markersize=3, color='red')
    # plt.plot(range(len_unet_plus), history_unet_plus['train_loss'].values, label='UNet cải tiến', linestyle='-', marker='x', markersize=3, color="#21f400")
    plt.title('Train Loss Comparison')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # Plotting Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(range(len_cae), history_cae['train_acc'].values, label='CAE', linestyle='-', marker='x', markersize=3)
    plt.plot(range(len_unet), history_unet['train_acc'].values, label='UNet', linestyle='-', marker='x', markersize=3, color='red')
    # plt.plot(range(len_unet_plus), history_unet_plus['train_acc'].values, label='UNet cải tiến', linestyle='-', marker='x', markersize=3, color="#21f400")
    plt.title('Train Accuracy Comparison')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
print(df_cae.shape[0])
plot_comparison_training_history(df_cae.shape[0], df_cae,
                                 df_unet.shape[0], df_unet,
                                 None, None)
```

```python
def plot_comparison_training_history(len_cae, history_cae, len_unet, history_unet, len_unet_plus, history_unet_plus, metric='mae'):

    plt.figure(figsize=(12, 6))
    print(f"Length of CAE history: {len_cae}")
    print(f"Length of UNet history: {len_unet}")
    print(f"Length of UNet Plus history: {len_unet_plus}")
    # Plotting MAE
    plt.subplot(1, 2, 1)
    plt.plot(range(len_cae), history_cae[f'train_{metric}'].values, label='CAE', linestyle='-', marker='x', markersize=3)
    plt.plot(range(len_unet), history_unet[f'train_{metric}'].values, label='UNet', linestyle='-', marker='x', markersize=3, color='red')
    # plt.plot(range(len_unet_plus), history_unet_plus[f'train_{metric}'].values, label='UNet cải tiến', linestyle='-', marker='x', markersize=3, color="#21f400")
    plt.title(f'Train {metric.upper()} Comparison')
    plt.xlabel('Epochs')
    plt.ylabel(metric.upper())
    plt.legend()
    plt.grid(True)

    # Plotting Validation MAE
    plt.subplot(1, 2, 2)
    plt.plot(range(len_cae), history_cae[f'val_{metric}'].values, label='CAE', linestyle='-', marker='x', markersize=3)
    plt.plot(range(len_unet), history_unet[f'val_{metric}'].values, label='UNet', linestyle='-', marker='x', markersize=3, color='red')
    # plt.plot(range(len_unet_plus), history_unet_plus[f'val_{metric}'].values, label='Unet cải tiến', linestyle='-', marker='x', markersize=3, color="#21f400")
    plt.title(f'Validation {metric.upper()} Comparison')
    plt.xlabel('Epochs')
    plt.ylabel(metric.upper())
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

metrics = ['mae', 'mse', 'rmse']
for metric in metrics:
    plot_comparison_training_history(df_cae.shape[0], df_cae,
                                     df_unet.shape[0], df_unet,
                                     None, None, metric=metric)
```
